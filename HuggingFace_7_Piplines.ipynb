```python
from transformers import BertTokenizer, TFBertModel,pipeline
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = TFBertModel.from_pretrained("bert-base-uncased")
text = "Replace me by any text you'd like."
encoded_input = tokenizer(text, return_tensors='tf')
output = model(encoded_input)
```

# 2) "Text Genration" Pipelines - uses an input prompt to genrate text

```python
from transformers import pipeline
classifier=pipeline('text-generation') # by default it take gpt2 model for text gerbation
classifier('In this course I will teach you how to ')
```

```python
# If you want to define whihc model to use , the you can tweak like
#https://huggingface.co/models?pipeline_tag=text-generation&sort=downloads - You can use various model of your choice form huggingface hub
from transformers import pipeline
classifier=pipeline('text-generation',model='distilgpt2')
classifier('In this course I will teach you how to ', max_length=200,num_return_sequences=2)

""" NOte:-
max_length=200, means maximum words in each sentences
num_return_sequences=2- means want to return 2 sentences
"""
```

3) "Fill Mask" Pipelines - will predict input word in the sentences

```python
from transformers import pipeline
classifier=pipeline('fill-mask')
classifier('The goal of life is to become man of <mask>.',top_k=2)

#tok_k=2 return top 2 results
```

#4) "NER" Name Entity Recognition Pipelines - identifies enitites such as person, organization or location in a sentences

```python
from transformers import pipeline
classifier=pipeline('ner', grouped_entities=True)
classifier('My Name is Sarthak and I work in Dassault Sys located in Pune')
```

#5) "question-answering" Pipelines - extract answers to a question from a given context

```python
from transformers import pipeline
question_answer=pipeline('question-answering')
question_answer(
    question='Where do I work?',
    context= 'My Name is Sarthak and I work in Halo in Pune'

)
```
